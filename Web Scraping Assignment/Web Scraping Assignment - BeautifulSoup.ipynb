I. Top 10 ODI teams in men’s cricket along with the records for matches, points and rating

```python
# To send requests to web page server
webpage = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')

# To scrape the Page Content
content = BeautifulSoup(webpage.content)

# Extracting the table
table = content.find('table', class_="table")
table.find_all('th')
 
# Extracting the Header
H = [i.text.strip('\n') for i in table.find_all('th')]
head = [i.split('\n')[0] for i in H]

# Extracting Data from the table body
data = []
rows = table.find_all('tr')[1:11]

for row in rows:
    co = row.find_all('td')
    cell = [ele.text.strip() for ele in co]
    data.append([ele for ele in cell if ele])

#cleaning unnecessary data
selected_data = [[i.split('\n')[0] for i in ele] for ele in data]

# Making DataFrame
top10_team = pd.DataFrame(data = selected_data, columns = head)
top10_team.drop(['Pos'], axis=1, inplace=True)
top10_team
```

Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and
make data frame

i) Headline

ii) Time

iii) News Link

```python
# Send the Request to the web page server
p = requests.get('https://www.cnbc.com/world/?region=world')

# To scrap the Page Content
s = BeautifulSoup(p.content)

# Articles in the website
articles = s.find_all('div', class_="LatestNews-headlineWrapper")

# Extracting Headlines from the article
Headlines = [i.find('a', class_="LatestNews-headline").text for i in articles]

# Extracting the links of the headlines
News_links = [i.find('a')['href'] for i in articles]

#extracting time
time = [i.find('time', class_="LatestNews-timestamp").text for i in articles]

#Making DataFrame
N = pd.DataFrame({"Headline": Headlines, "Time":time, "links":News_links})
N
```

Write a python program to scrape the details of most downloaded articles from AI in last 90
days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles
Scrape below mentioned details and make data frame

i) Paper Title
ii) Authors
iii) Published Date
iv) Paper URL

```python
# Send the Request to the web page server
P = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')

# To scrap the Page Content
soup = BeautifulSoup(P.content)

#Extracting the Titles
paper_title = [i.text for i in soup.find_all('a',class_="sc-5smygv-0 fIXTHm")]

#Extracting the Author Name
authors = [i.text for i in soup.find_all('span', class_="sc-1w3fpd7-0 dnCnAO")]
authors

#extracting Published date
published_date = [i.text for i in soup.find_all('span',class_="sc-1thf9ly-2 dvggWt")]

# Extracting All Paper links
paper_URL = [i.get('href') for i in soup.find_all('a', class_="sc-5smygv-0 fIXTHm")]

#Making DataFrame
df = pd.DataFrame({'Paper Title':paper_title, 'Author':authors, 'Published Date':published_date, 'Paper URL':paper_URL})
df
```

Write a python program to scrape mentioned details from dineout.co.inand make data framei) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
# Send the Request
url = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')

# To scrap the Page Content
s = BeautifulSoup(url.content)

# 1. To Scrap the Restaurant name
restaurant_names = [i.text for i in s.find_all('a',class_="restnt-name ellipsis")]

# 2. To Scrap the cuisine
cuisine = [i.text.split('|')[1] for i in s.find_all('span', class_="double-line-ellipsis")]

# 3. To Scrap the Location
location = [i.text for i in s.find_all('div', class_="restnt-loc ellipsis")]
location

# 4. To Scrap the Ratings
ratings = [i.text for i in s.find_all('div', class_="restnt-rating rating-4")]


# 5. To Scrap the Image Urls
images = [i.get('data-src') for i in s.find_all('img', class_="no-img")]

# TO make dataframe
df = pd.DataFrame({'Restaurant Names':restaurant_names, 'Cuisine':cuisine, 'Location':location, 'Rating':ratings,'Images':images })
df
```
